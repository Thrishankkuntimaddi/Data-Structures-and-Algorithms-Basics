# Binary search

I/P : [10, 20, 30, 40, 50]

x = 2

O/P = 1

```python
# Naive Solution

def binary(l,x):
  low = 0
  high = len(l) - 1
  while low <= high:
    mid = (low + high) // 2
    if l[mid] == x:
      return mid
    elif l[mid] < x:
      low = mid + 1
    else:
      high = mid - 1

  return -1

l = [10, 20, 30, 40, 50, 60]
x = 40
print(binary(l,x))
```

```python
# Recursive Binary Search

def binary(l, x, low, high):
  if low > high:
    return -1
  mid = (low + high) // 2
  if l[mid] == x:
    return mid
  elif l[mid] < x:
    return binary(l, x, mid + 1, high)
  else:
    return binary(l, x, low, mid - 1)

l = [10, 20, 30, 40, 50, 60]
x = 50

print(binary(l, x, 0, len(l) - 1))
```

# Bineay Search for Unsorted List

```python
# Method 1

aux_space = []

def make_aux(arr, n):
  global aux_space
  for i in range(n):
    aux_space.append(arr[i])

  aux_space.sort()

l = [10, 20, 30, 40, 50, 60, 9, 10, 22, 54, 21, 29]
make_aux(l, len(l))
print(binary(aux_space, 9, 0, len(aux_space) - 1))
print(aux_space)
```

```python
# Method 2

def binary_search(arr, target):
    left, right = 0, len(arr) - 1

    while left <= right:
        mid = left + (right - left) // 2 # avoiding the Overflow
        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            left = mid + 1
        else:
            right = mid - 1
    return -1

arr = [3, 6, 1, 9, 4, 2]
arr.sort()
res = binary_search(arr, 3)
print(res)
```

```python
# Method 3

def findPeak(nums):
  left, right = 0 ,len(nums) - 1

  while left < right:
    mid = left + (right - left) // 2
    if nums[mid] < nums[mid + 1]:
      left = mid + 1
    else:
      right = mid
  return left

nums = [1, 2, 1, 3, 5, 6, 4]
print(findPeak(nums))
```

# Index of First Occurrence : Sorted Order

I/P : arr[] = [1, 10, 20, 10, 20, 30, 30, 10 ]

x = 20

O/P : 2



# For the first occurrence of a number

1. If (high >= low)

2. Calculate  mid = low + (high - low)/2;

3. If ((mid == 0 || x > arr[mid-1]) && arr[mid] == x)  : return mid;

4. Else if (x > arr[mid])  : return first(arr, (mid + 1), high, x, n);

5. Else  : return first(arr, low, (mid -1), x, n);

6. Otherwise return -1;

```python
# Naive : Unsorted & Sorted Order

def firstOcc(arr, n, x):
  for i in range(0, n):
    if arr[i] == x:
      return i
  return - 1

l = [1, 10, 20, 10, 20, 30, 30, 10]
n = len(l)
x = 20
print(firstOcc(l, n, x))

# Time Complexity : O(n)
# Auxiliary Space : O(1)
```

```python
# Efficient : Using Recursion

def firstOcc(arr, low, high, x, n):
  if low > high:
    return -1

  if(high >= low):
    mid = low + (high - low) // 2
    if((mid == 0 or x > arr[mid - 1]) and arr[mid] == x):
      return mid
    elif(x > arr[mid]):
      return firstOcc(arr, (mid + 1), high, x, n)
    else:
      return firstOcc(arr, low, (mid - 1), x, n)

  return -1

l = [10, 10, 20, 30, 40, 40, 40, 40]

n = len(l)
x = 20
print(firstOcc(l, 0, n - 1, x, n))

# Time Complexity : O(logn)
# Auxiliary Space : O(n)
```

```python
# Efficient using Iteration

def first(arr, n, x):
  low = 0
  high = n - 1
  while low <= high:
    mid = (low + high) // 2
    if x > arr[mid]:
      low = mid + 1
    elif x < arr[mid]:
      high = mid - 1
    else:
      if mid == 0 or x > arr[mid - 1]:
        return mid
      else:
        high = mid - 1

  return -1

l = [10, 10, 20, 30, 40, 40, 40, 40]
n = len(l)
x = 20
print(first(l, n, x))

# Time Complexity : O(log n)
# Space Complexity : O(1)
```

# Index of Last Occurrence : Sorted Order

I/P : arr[] = [1, 10, 20, 10, 20, 30, 30, 10 ]

x = 20

O/P : 4


# For the Last occurrence of a number

1. if (high >= low)
2. calculate mid = low + (high - low)/2;
3. if( ( mid == n-1 || x < arr[mid+1]) && arr[mid] == x ) : return mid;
4. else if(x < arr[mid]) : return last(arr, low, (mid -1), x, n);
5. else : return last(arr, (mid + 1), high, x, n);      
6. otherwise return -1;



```python
# Naive : Unsorted & Sorted Orders

def findLastOccurrence(arr, n, x):
    last = -1
    for i in range(n):
        if arr[i] == x:
            last = i

    if last != -1:
        print("Last Occurrence = ", last)
    else:
        print("Not Found")

l = [10, 20, 30, 20, 30, 40, 40, 40]
n = len(l)
x = 30
findLastOccurrence(l, n, x)

# Time Complexity : O(n)
# Auxiliary Space : O(1)
```

```python
# Efficient : Using Recursion

def last(arr, low, high, x, n):
	if (high >= low):
		mid = low + (high - low) // 2
		if ((mid == n - 1 or x < arr[mid + 1]) and arr[mid] == x):
			return mid
		elif (x < arr[mid]):
			return last(arr, low, (mid - 1), x, n)
		else:
			return last(arr, (mid + 1), high, x, n)

	return -1


arr = [1, 2, 2, 2, 2, 3, 4, 7, 8, 8]
n = len(arr)

x = 8
print("Last Occurrence = ", last(arr, 0, n - 1, x, n))

# Time Complexity : O(log n)
# Auxiliary Space : O(n)
```

# Count Occurences in a Sorted Array

I/P : l = [10, 20, 20, 20, 30, 30]

x = 20

O/P : 3

NOTE :

Naive Solution : Linear Traversal : O(n)

Better Solution : Simple Binary search : O(logn + K)

```python
# Better Solution

def search(nums, target, find_start_index):
    answer = -1
    start = 0
    end = len(nums) - 1
    while start <= end:
        mid = start + (end - start) // 2
        if target < nums[mid]:
            end = mid - 1
        elif target > nums[mid]:
            start = mid + 1
        else:
            answer = mid
            if find_start_index:
                end = mid - 1
            else:
                start = mid + 1
    return answer

def count_occurrences(nums, x):
    first_index = search(nums, x, True)
    last_index = search(nums, x, False)
    if first_index != -1 and last_index != -1:
        return last_index - first_index + 1
    else:
        return 0

if __name__ == "__main__":
    arr = [1, 2, 2, 2, 2, 3, 4, 7, 8, 8]
    x = 2
    c = count_occurrences(arr, x)
    print(c)


# Time Complexity : O(log n)
# Space Complexity: O(1)

```

```python
# Efficient

def firstOcc(arr, x):
  n = len(arr)
  for i in range(0, n):
    if arr[i] == x:
      return i
  return - 1

def lastOcc(arr, x):
  n = len(arr)
  for i in range(n - 1, -1, -1):
    if arr[i] == x:
      return i
  return -1

def countOcc(l, x):
  first = firstOcc(l, x)
  if first == -1:
    return 0
  else:
    return lastOcc(l, x) - first + 1

l = [10, 20, 20, 20, 20, 20, 20, 30, 30]
x = 20
print(countOcc(l, x))

# Time Complexity : O(logn)
# Auxiliary Space : O(n)
```

# Count 1's in a Sorted Binary List

Input: arr[] = {1, 1, 0, 0, 0, 0, 0}

Output: 2

Input: arr[] = {1, 1, 1, 1, 1, 1, 1}

Output: 7

Input: arr[] = {0, 0, 0, 0, 0, 0, 0}

Output: 0

# Naive approach:

  A simple solution is to linearly traverse the array until we find the 1's in the array and keep count of 1s. If the array element becomes 0 then return the count of 1's.

Time Complexity: O(n)

Auxiliary Space: O(1)


# Efficient : Count 1's in a sorted binary array using Binary search recursively

Do while low <= high:
 - Calculate mid using low + (high - low) / 2.
 - Check if the element at mid index is the last 1
 - If the element is not last 1, move the low to right side recursively and return the result received from it.
 - Otherwise, move the low to left recursively and return the result received from it.


```python
def countOnes(arr, low, high):
    if high >= low:
        mid = low + (high-low)//2
        if ((mid == high or arr[mid+1] == 0) and (arr[mid] == 1)):
            return mid+1
        if arr[mid] == 1:
            return countOnes(arr, (mid+1), high)
        return countOnes(arr, low, mid-1)

    return 0

arr = [1, 1, 1, 1, 0, 0, 0]
print("Count of 1's in given array is", countOnes(arr, 0, len(arr)-1))

# Time complexity: O(log(n))
# Auxiliary Space: O(log(n))
```

# Efficient : Count 1's in a sorted binary array using binary search iteratively

Do while low <= high:
 - Calculate the middle index say mid
 - Check if arr[mid] is less than 1 then move the high to left side (i.e, hight = mid - 1)
 - If the element is not last 1 then move the low to the right side (i.e, low = mid + 1)
 - Check if the element at the middle index is last 1 then return mid + 1
Otherwise move to low to right (i.e, low = mid + 1)

```python
def countOnes(arr, n):
    low = 0
    high = n - 1
    while (low <= high):
        mid = (low + high) // 2
        if ((mid == n - 1 or arr[mid + 1] == 0) and arr[mid] == 1):
            return mid + 1
        if (arr[mid] == 0):
            high = mid - 1
        elif(arr[mid] == 1):
            low = mid + 1

    return 0


if __name__ == '__main__':
    arr = [1, 1, 1, 1, 0, 0, 0]
    n = len(arr)
    print("Count of 1's in given array is ", countOnes(arr, n))

# Time complexity: O(log(n))
# Auxiliary Space: O(1)
```

# Square Root

I/P : x = 4

O/P : 2

```python
# Naive

def sqrtt(x):
  i = 1
  while i*i <= x:
    i += 1
  return i - 1

print(sqrtt(4))

# Time Complexity : O(sqrt(x))
# Auxiliary Space : O(1)
```

```python
# Efficient

def sqrt(x):
  low = 1
  high = x
  ans = -1

  while low <= high:
    mid = (low + high) // 2
    msq = mid * mid
    if msq == x:
      return mid
    elif msq < x:
      low = mid + 1
      ans = mid
    else:
      high = mid - 1
  return ans

print(sqrt(4))


# Time Complexity : O(logn)
# Space Complexity : O(1)
```
